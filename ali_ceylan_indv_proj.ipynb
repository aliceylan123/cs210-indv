```python
import pandas as pd
from os.path import join
filename = "taxi-trips.csv"
d_path = "./data"
myTable = pd.read_csv(join(d_path, filename))
myTable
```

```python
myTable['pickup_district']=""
myTable['dropoff_district']=""
myTable['distance']=0.0
myTable
```

```python
import reverse_geocoder as rg
from geopy.distance import geodesic
pickup_coords=list()
dropoff_coords=list()
for row in myTable.itertuples():
    pickup_coords.append((row.pickup_latitude,row.pickup_longitude))
    dropoff_coords.append((row.dropoff_latitude,row.dropoff_longitude))
pickup_results=rg.search(pickup_coords)
dropoff_results=rg.search(dropoff_coords)
for row in myTable.itertuples():
    myTable.set_value(row.Index,'pickup_district',pickup_results[row.Index]['name'])
    myTable.set_value(row.Index,'dropoff_district',dropoff_results[row.Index]['name'])
    myTable.set_value(row.Index,'distance',geodesic(pickup_coords[row.Index],dropoff_coords[row.Index]).m)
myTable
```

```python
preferred_pickup=myTable['pickup_district'].value_counts()
preferred_dropoff=myTable['dropoff_district'].value_counts()
preferred_pickup=preferred_pickup[:5]
preferred_pickup

```

```python
preferred_dropoff=preferred_dropoff[:5]
preferred_dropoff
```

```python
import datetime
from datetime import datetime as dt
from datetime import time
myTable['time_of_day']="A"
seven=datetime.time(7,0,0)
nine=datetime.time(9,0,0)
four=datetime.time(16,0,0)
six=datetime.time(18,0,0)
eleven=datetime.time(23,0,0)
for row in myTable.itertuples():
    mydate=dt.strptime(row.pickup_datetime,"%Y-%m-%d %H:%M:%S")
    if mydate.time() >=seven and mydate.time()<nine:
        myTable.set_value(row.Index,'time_of_day',"rush_hour_morning")
    elif mydate.time() >=nine and mydate.time()<four:
        myTable.set_value(row.Index,'time_of_day',"afternoon")
    elif mydate.time() >=four and mydate.time()<six:
        myTable.set_value(row.Index,'time_of_day',"rush_hour_evening")
    elif mydate.time() >=six and mydate.time()<eleven:
        myTable.set_value(row.Index,'time_of_day',"evening")
    else:
        myTable.set_value(row.Index,'time_of_day',"late_night")
myTable
```

```python
import matplotlib.pyplot as plt
twocols=myTable[['distance','time_of_day']]
dot_group = twocols.groupby(by="time_of_day").mean()

ax = dot_group.plot(kind="bar",
                   color="steelblue",
                   rot=70)


plt.ylabel("Avg. Distance")
plt.title("Avg Distance vs Day of Time")
plt.show()
```

```python
threecols=myTable[['pickup_datetime','dropoff_datetime','time_of_day']]
threecols['duration']=0
for row in threecols.itertuples():
    dest=dt.strptime(row.dropoff_datetime,"%Y-%m-%d %H:%M:%S")
    source=dt.strptime(row.pickup_datetime,"%Y-%m-%d %H:%M:%S")
    duration=dest-source
    threecols.set_value(row.Index,'duration',duration.total_seconds()/60.0)
seperatedataframe=threecols[['duration','time_of_day']]
dot_group = seperatedataframe.groupby(by="time_of_day").mean()

ax = dot_group.plot(kind="bar",
                   color="steelblue",
                   rot=70)


plt.ylabel("Avg. Duration (mins)")
plt.title("Avg Duration vs Day of Time")
plt.show()
```

```python
size = pd.crosstab(myTable["passenger_count"], myTable["distance"])
size
```

```python
from scipy import stats
stats.chi2_contingency(size)
print("p-value is 1.0. So there is no significant data that shows passenger_count and distance information are related to each other")
```

```python
day_of_week_table=myTable[['pickup_datetime','distance']]
day_of_week_table['day_of_week']=0
for row in myTable.itertuples():
    mydate=dt.strptime(row.pickup_datetime,"%Y-%m-%d %H:%M:%S")
    day_of_week_table.set_value(row.Index,'day_of_week',mydate.weekday())
days=pd.crosstab(day_of_week_table['day_of_week'],day_of_week_table['distance'])
days
```

```python
stats.chi2_contingency(days)
print("p-value is 0.495. So there is no significant data that shows day of the week and distance information are related to each other")
```
